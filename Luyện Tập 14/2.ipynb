```python
n = int(input("n: "))

def h1(n):
    for i in range(1, n+1):
        print("*" * i)
def h2(n):
    for i in range(1, n+1):
        print(" " * (n-i) + "*" * i)
def h3(n):
    for i in range(n, 0, -1):
        print("*" * i)
def h4(n):
    for i in range(n, 0, -1):
        print(" " * (n-i) + "*" * i)
def h5(n):
    for i in range(1, n+1):
        print(" " * (n-i) + "*" * (2*i-1))
def h6(n):
    for i in range(n, 0, -1):
        print(" " * (n-i) + "*" * (2*i-1))
def h7(n):
    for i in range(1, n+1):
        print(" " * (n-i) + "*" * (2*i-1))
    for i in range(n-1, 0, -1):
        print(" " * (n-i) + "*" * (2*i-1))
def h8(n):
    for i in range(n):
        print(" " * i + "*")
def h9(n):
    for i in range(n):
        print(" " * (n-i-1) + "*")
def h10(n):
    for i in range(n):
        line = ""
        for j in range(n):
            if j == i or j == n-i-1:
                line += "*"
            else:
                line += " "
        print(line)
print("1:")
h1(n)
print("\n2:")
h2(n)
print("\n3:")
h3(n)
print("\n4:")
h4(n)
print("\n5:")
h5(n)
print("\n6:")
h6(n)
print("\n7:")
h7(n)
print("\n8:")
h8(n)
print("\n9:")
h9(n)
```
