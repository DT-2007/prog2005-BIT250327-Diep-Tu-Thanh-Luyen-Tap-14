```python
n = int(input("n: "))

def h1(n):
    for i in range(n):
        print(" ".join(["1"] * n))
def h2(n):
    for i in range(n):
        print(" ".join([str(j+1) for j in range(n)]))
def h3(n):
    for i in range(1, n+1):
        print(" ".join([str(j) for j in range(1, i+1)]))
def h4(n):
    for i in range(n, 0, -1):
        print(" ".join([str(j) for j in range(1, i+1)]))
def h5(n):
    for i in range(1, n+1):
        print(" ".join([str(j) for j in range(1, i+1)]))
def h6(n):
    for i in range(1, n+1):
        row = []
        for j in range(1, n+1):
            if i == 1 or j == 1 or i == j or j == n:
                row.append(str(j))
            else:
                row.append(" ")
        print(" ".join(row))
def h7(n):
    for i in range(1, n+1):
        print(" ".join([str(i)] * i))
def h8(n):
    for i in range(1, n+1):
        row = [str(j) for j in range(1, i+1)]
        row = row + row[-2::-1]
        print(" " * (n-i) * 2 + " ".join(row))
def h9(n):
    for i in range(1, n+1):
        if i == 1:
            print(" " * (n-i) * 2 + "1")
        elif i < n:
            print(" " * (n-i) * 2 + "1" + " " * (2*i-3) * 2 + "1")
        else:
            row = [str(j) for j in range(1, i+1)]
            row = row + row[-2::-1]
            print(" ".join(row))
print("1:")
h1(n)
print("\n2:")
h2(n)
print("\n3:")
h3(n)
print("\n4:")
h4(n)
print("\n5:")
h5(n)
print("\n6:")
h6(n)
print("\n7:")
h7(n)
print("\n8:")
h8(n)
print("\n9:")
h9(n)
```
